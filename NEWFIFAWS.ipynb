## Work on the contract column

```python
df.Contract.unique()
```

### * Observation: some players are on contract while some are on loan, in other to make the column type uniform, we give players with active contract as active while the ones on loan should be categorized as on loan

```python
# define a function to change the values with 'on loan' as an "On loan" category
# define the players with active loans as Active and they can be located easily cause they contain '~'
def contract(a):
    if "On Loan" in a:
        a = 'ON LOAN'
        return a
    elif '~' in a:
        a = 'ACTIVE'
        return a
    elif 'Free' in a:
        a = "FREE"
        return a
```

```python
df.Contract = df.Contract.apply(contract)
```

```python
df.Contract.unique()
```

```python
df.Contract = df.Contract.astype('category')
```

```python
# Rename the column to Contract status
df = df.rename(columns={'Contract':'ContractStatus'})
```

## Check out the Hits column

```python
df.Hits.unique()
```

### * Observation: some values have 'k' in them which signify they are in thousand, change the value to its actual numeric status

```python
# define a function to replace K with an empty string and turn the value to a float so it can be added to 1000
def hitchange(h):
    if "K" in h:
        h = h.replace('K', "")
        h = float(h)* 1000
        return round(h)
    else:
        return h
```

```python
df.Hits = df.Hits.apply(hitchange)
```

```python
df.Hits.unique()
```

## From the observation we have to change the height and weight column 

```python
df.Height.unique()
```

```python
def convert_height(x):
    if x[-1] == '"':
        x = x.replace("\"","")
        inch = int(x[2:]) * 2.54
        foot = int(x[0]) * 30.48
        return round(foot+inch)
    elif x[-1] == "m":
        return int(x[:-2])
```

```python
df.Height = df.Height.apply(convert_height)
```

```python
df.Height.unique()
```

### Rename the column name to Height_cm

```python
df = df.rename(columns={'Height':'Height_cm'})
```

```python
def weight_kg(x):
    if 'lbs' in x:
        x = x.replace('lbs','')
        x = float(x) * 0.45359237
        x = int(x)
        return x
    else:
        return int(x[:-2])
```

```python
df.Weight = df.Weight.apply(weight_kg)
```

## Work on the Release clause, Wage and Value columns to suit their respective positions 

### RELEASE CLAUSE

```python
df["Release Clause"].unique()
```

```python
def release(x):
    if 'M' in x:
        x = x.replace('M', '')
        x = float(x[1:]) * 1000000
        return round(x)
    elif "K" in x:
        x = x.replace('K', '')
        x = float(x[1:]) * 1000
        return round(x)
    else:
        return x
```

```python
df["Release Clause"] = df["Release Clause"].apply(release)
```

```python
df["Release Clause"].unique()
```

```python
# Rename Release Clause to have the Euro sign
```

```python
df = df.rename(columns={'Release Clause':'Release_Clause €'})
```

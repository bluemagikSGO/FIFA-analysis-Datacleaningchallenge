## some values in the Hit columns contain 'K', so we change those values to thousand

```python
result = df[df['Hits'].str.contains('K')]
```

```python
result
```

```python
# Remove the "." and replace "K" to 00
```

```python
remove_ch = lambda x: x.replace('.', '').replace('K', '00') if isinstance(x, str) else x
```

```python
result.Hits = result.Hits.apply(remove_ch)
```

```python
result.Hits.reset_index()
```

### Some of the thousands came out to be hundred so we change those hundreds to thousands

```python
index_list = [416,690,8116,15819]
```

```python
rows_to_change = result.loc[index_list]
```

```python
rows_to_change['Hits'] = rows_to_change['Hits'].apply(lambda x: x.replace('00', '000'))
```

### After updating the hundreds to thousands, update the output to the result variable which hold values that were formerly 'k's

```python
result.update(rows_to_change)
```

```python
result
```

```python
df.update(result)
```

```python
df
```

```python
df.Hits = df.Hits.astype("int64")
```

## Work on the height and weight column

# Height Cleaning

```python
df.Height.unique()
```

```python
#There are some feet in the height column that is supposed to be in centimeter, we have to get those values that are in feet
feet = df[df['Height'].str.contains('"')]
```

```python
feet
```

```python
# Remove the characters that makes the values a feet 
replace = lambda x: x.replace("'",".").replace('"',' ')
```

```python
feet['Height'] = feet['Height'].apply(replace).astype(float)
```

```python
#Using the lambda function, change the feets to centimeter
rp = lambda x: x * 30.48
feet['Height'] = feet['Height'].apply(rp).astype('int64')
```

```python
feet
```

```python
feet['Height'] = feet['Height'].astype('str')
```

```python
feet['Height']
```

```python
# Add the suffix 'cm' to make the values match those in the main dataframe
```

```python
func = lambda x: x + "cm"
```

```python
feet.Height = feet["Height"].apply(func)
```

```python
feet
```

```python
df.update(feet)
```

# Weight Cleaning

```python
df.head(2)
```

## create a df for the weights measured in lbs

```python
lbs = df[df['Weight'].str.contains('lbs')]
```

```python
lbs
```

## Remove the lbs

```python
change = lambda x: x.replace("lbs"," ")
```

```python
lbs.Weight = lbs.Weight.apply(change)
```

```python
# change the values to float values so we can add it up to kilogram
lbs.Weight = lbs.Weight.astype(float)
```

```python
# add up with the kg value to turn lbs to kilogram
lbs_to_kg = lambda x: x / 2.2046
```

```python
lbs.Weight = lbs.Weight.apply(lbs_to_kg).astype('int')
```

```python
#change from 
lbs.Weight = lbs.Weight.astype('str')
```

```python
new_kg = lambda x: x + 'kg'
```

```python
lbs.Weight = lbs.Weight.apply(new_kg)
```

## some values in the Hit columns contain 'K', so we change those values to thousand

```python
result = df[df['Hits'].str.contains('K')]
```

```python
result
```

```python
remove_ch = lambda x: x.replace('.', '').replace('K', '00') if isinstance(x, str) else x
```

```python
result.Hits = result.Hits.apply(remove_ch)
```

```python
result.Hits.reset_index()
```

### Some of the thousands came out to be hundred so we change those hundreds to thousands

```python
index_list = [416,690,8116,15819]
```

```python
rows_to_change = result.loc[index_list]
```

```python
rows_to_change['Hits'] = rows_to_change['Hits'].apply(lambda x: x.replace('00', '000'))
```

### After updating the hundreds to thousands, update the output to the result variable which hold values that were formerly 'k's

```python
result.update(rows_to_change)
```
